```python
import pandas as pd
```

```python
df = pd.read_csv('../data/processed/2016-2019_3std_preprocessed.csv')
df.head(9)
```

```python
df['AGU'].isnull().sum()/len(df['AGU'])
```

```python
est = ['AGU','ATM','CEN','LDO','LPIN','MIR','OBL','SFE','TLA','VAL']
```

```python
for e in est:
    n = df[e].isnull().sum()/len(df[e])
    p = df.loc[df['PARAM']=='PM10'][e].isnull().sum() / len(df.loc[df['PARAM']=='PM10'][e])
    print(e,n,'PM10',p)
```

```python
df_pin = df.drop(columns=['Unnamed: 0'])
df_pin = df_pin.loc[df_pin.PARAM=='PM10']
df_pin.drop(columns=['PARAM'], inplace=True)
df_pin.set_index(['FECHA','HORA'], inplace=True)
df_pin.head(9)
```

```python
names = []
new_n = {}
for c in df_pin.columns:
    new_n[c] = c + '_PM10'
new_n
```

```python
df_pin.rename(columns=new_n, inplace=True)
df_pin.head()
```

```python
df_meteo = df[df['PARAM'].isin(['TMP','HR','WS'])].copy()
df_meteo = df_meteo[['FECHA','HORA','PARAM','LPIN']]
df_meteo.set_index(['FECHA','HORA'], inplace=True)
df_meteo.head()
```

```python
df_meteo = df_meteo.pivot(columns='PARAM').copy()
df_meteo.head()
```

```python
df_meteo.columns = df_meteo.columns.droplevel()
df_meteo.columns
```

```python
df_merge = pd.merge(df_pin, df_meteo, left_index=True, right_index=True)
df_merge
```

```python
df_merge['LPIN_PM10'].iloc[1:-1]
```
