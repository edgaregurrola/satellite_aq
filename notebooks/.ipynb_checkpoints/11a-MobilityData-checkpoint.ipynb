```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/external/Viajes_BRT/Viajes_BRT_2019.csv', encoding='iso-8859-1')
print(df.shape)
df.head(5)
```

```python
df.ESTACION.unique()
```

Media por día ya que los datos de personas ocupadas, y personas viviendo no tienen temporalidad, se puede asumir que los otros también manejan información en un día promedio, por lo tanto no afectan las unidades.

```python
df['FECHA'] = pd.to_datetime(df['FECHA'])
df.head(2)
```

```python
df_tmp = df.groupby(['FECHA','NO_STA','ESTACION']).sum().copy()
df_tmp
```

```python
df_group = df_tmp.reset_index().groupby('ESTACION').mean()
df_group
```

```python
df_group.to_csv('../output/data/Mobility_BRT/Viajes_BRT_PromedioDiario.csv')
```
